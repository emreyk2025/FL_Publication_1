```python
# Import dependencies
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the simulation metrics csv file for results
df = pd.read_csv("simulation_metrics_second_attempt.csv")
```

Results with number of meters 50, 100 and all meters (4411)

```python
print(df)
```

```python
# Filter the data to keep everything else equal and look at one aspect of the results at a time

filtered_data_1 = df[(df['model_name'].isin(['LSTM_Simple'])) & 
                   (df['neuron_multiplier'] == 2) & 
                   (df['loss_function'] == 'MSE') & 
                   (df['data_dropout'] == 0.0)]
```

```python
print(filtered_data_1)
```

```python
filtered_data_2 = df[(df['model_name'].isin(['LSTM_stacked'])) & 
                   (df['neuron_multiplier'] == 2) & 
                   (df['loss_function'] == 'MSE') & 
                   (df['data_dropout'] == 0.0)]
```

```python
print(filtered_data_2)
```

```python
# See the training time with number of meters:

# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: First dataset
sns.barplot(data=filtered_data_1, x='num_meters', y='training_time', ax=axes[0])
axes[0].set_title('Training time per Number of Meters for LSTM_Simple Model\nwith 2x Neurons and No Data Dropout', 
                  fontsize=12)
axes[0].set_xlabel('Number of Meters')
axes[0].set_ylabel('Training Time')
axes[0].set_ylim(0, 40000)


# Plot 2: Second dataset
sns.barplot(data=filtered_data_2, x='num_meters', y='training_time', ax=axes[1])
axes[1].set_title('Training time per Number of Meters for LSTM_stacked Model\nwith 2x Neurons and No Data Dropout', 
                  fontsize=12)
axes[1].set_xlabel('Number of Meters')
axes[1].set_ylabel('Training Time')
axes[1].set_ylim(0, 40000)


# Adjust layout to prevent overlap
plt.tight_layout()
```

```python
# Create a figure with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: First dataset
sns.barplot(data=filtered_data_1, x='num_meters', y='mae', ax=axes[0])
axes[0].set_title('MAE per Number of Meters for LSTM_Simple Model\nwith 2x Neurons and No Data Dropout', 
                  fontsize=12)
axes[0].set_xlabel('Number of Meters')
axes[0].set_ylabel('MAE')
axes[0].set_ylim(0.35, 0.52)


# Plot 2: Second dataset
sns.barplot(data=filtered_data_2, x='num_meters', y='mae', ax=axes[1])
axes[1].set_title('MAE per Number of Meters for LSTM_stacked Model\nwith 2x Neurons and No Data Dropout', 
                  fontsize=12)
axes[1].set_xlabel('Number of Meters')
axes[1].set_ylabel('MAE')
axes[1].set_ylim(0.35, 0.52)


# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```
